```python
from os.path import join
import yaml
import synonymizer.synonymize as syn
import pandas as pd
```

```python
SYNONYM_RULES = "../rulebook/synonym_rules.yaml"
SCHEMA = "../schema.yaml"
TEST_DATA_FOLDER = "../tests/data/"
envo_termlist = join(TEST_DATA_FOLDER,"envo_termlist.tsv")
go_termlist = join(TEST_DATA_FOLDER,"go_termlist.tsv")

envo_op = join(TEST_DATA_FOLDER, "envo_syn_termlist.tsv")
go_op = join(TEST_DATA_FOLDER, "go_syn_termlist.tsv")

```

```python
syn.run(rule_file=SYNONYM_RULES, schema_file=SCHEMA, data_folder= TEST_DATA_FOLDER )
```

```python
envo_input_df = pd.read_csv(envo_termlist, sep='\t', header=None)
envo_input_df.sort_values(by=[2], inplace=True, ascending=False, ignore_index=True)
envo_input_df
```

```python
envo_synonymized_df = pd.read_csv(envo_op, sep='\t', header=None)
envo_synonymized_df.sort_values(by=[2], inplace=True, ascending=False, ignore_index=True)
envo_synonymized_df
```

```python
go_input_df = pd.read_csv(go_termlist, sep='\t', header=None)
go_input_df.sort_values(by=[2], inplace=True, ascending=False, ignore_index=True)
go_input_df
```

```python
go_synonymized_df = pd.read_csv(go_op, sep='\t', header=None)
go_synonymized_df.sort_values(by=[2], inplace=True, ascending=False, ignore_index=True)
go_synonymized_df
```
